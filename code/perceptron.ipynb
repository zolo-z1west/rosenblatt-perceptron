ok so the plan is to show the inputs and weights as list. 
rosenblatt's paper basically shows a artificial neuron with several inputs and their respective weights along with bias variable.

we will first define a 'summation' function which implements the summation rule as stated in the paper -

```python
def summation(x, w, b):
    total = 0           # this is our output 
    for i in range(len(x)):
        total += x[i] * w [i]
    total += b                # adding bias to the output
    return total 
```
